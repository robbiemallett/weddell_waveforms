```python
import requests
import tqdm
from bs4 import BeautifulSoup
import sys
import pickle
import dill
```

```python
exists=[]
for i in tqdm.trange(1,80_000):
    
    response = requests.get(f'https://mdpi.com/si/{i}')
    if response.status_code == 200:
        exists.append(True)
    else:
        exists.append(False)
        
    break
```

```python
soups = []
titles = []

for i in tqdm.trange(10,15):
    
    reqs = requests.get(f'https://mdpi.com/si/{i}')
    if reqs.status_code == 200:
        soup = BeautifulSoup(reqs.text, 'html.parser')
        for title in soup.find_all('title'):
            t = title.get_text()
            titles.append(t)
            print(t)
        articles=0
        reviews=0
        for doc in soup.find_all('span',class_='label articletype'):
            if 'Article' in str(doc):articles+=1
            if 'Review' in str(doc):reviews+=1
                
        for doc in soup.find_all('span',class_='sciprofiles-link__name'):
#             print(doc)
            print(str(doc).split('>')[-2].split('<')[0])
    
    break

    
#         break
```
